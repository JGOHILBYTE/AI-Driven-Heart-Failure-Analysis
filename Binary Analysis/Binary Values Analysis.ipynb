```python

bar_width = 0.35


r1 = np.arange(len(gender_survivors_binary_df["BINARY CATEGORIES"]))
r2 = [x + bar_width for x in r1]


fig, ax = plt.subplots()


bars1 = ax.bar(r1, gender_survivors_binary_df["MALE_COUNT"], color='blue', width=bar_width, edgecolor='grey', label='Male')
bars2 = ax.bar(r2, gender_survivors_binary_df["FEMALE_COUNT"], color='pink', width=bar_width, edgecolor='grey', label='Female')


ax.set_xlabel('Categories', fontweight='bold')
ax.set_ylabel('Count', fontweight='bold')
ax.set_title('Male and Female Binary Counts for Survivors')
ax.set_xticks([r + bar_width / 2 for r in range(len(gender_survivors_binary_df["BINARY CATEGORIES"]))])
ax.set_xticklabels(gender_survivors_binary_df["BINARY CATEGORIES"])

for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center')

plt.grid()


ax.legend()


plt.show()
```
