```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Analysis of Male Survivors


```python
males = pd.read_csv("MaleSurvivor.csv")
```

```python
males
```

```python
males.describe()
```

## Analysis of CPK


```python
max(males["creatinine_phosphokinase"])
```

```python
males["CPK Levels"] = pd.cut(x=males["creatinine_phosphokinase"],bins = [10,120,7861], labels=["normal","high"])
```

```python
males["CPK Levels"]
```

```python
print(males["CPK Levels"].value_counts())
```

```python
counts = males['CPK Levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of CPK Levels")
plt.xlabel("CPK Levels")
plt.ylabel("Count")
plt.show()
```

```python
males["ejection fraction levels"] = pd.cut(x=males["ejection_fraction"],bins = [0,30,40,51,72], labels=["severely abnormal","moderately abnormal","mildly abnormal","normal"])
```

```python
males["ejection fraction levels"]
```

```python
counts = males['ejection fraction levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of ejection fraction levels")
plt.xlabel("ejection fraction levels")
plt.ylabel("Count")
plt.show()
```

```python
max(males["serum_creatinine"])
```

```python
males["serum creatinine levels"] = pd.cut(x=males["serum_creatinine"],bins = [0,0.7,1.3,6.1], labels=["low","normal","abnormal"])
```

```python
males["serum creatinine levels"]
```

```python
counts = males['serum creatinine levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of serum creatinine levels")
plt.xlabel("serum creatinine levels")
plt.ylabel("Count")
plt.show()
```

```python
max(males["serum_sodium"])
```

```python
males["serum sodium levels"] = pd.cut(x=males["serum_sodium"],bins = [0,135,145,148], labels=["low","normal","abnormal"])
```

```python
males["serum sodium levels"]
```

```python
counts = males['serum sodium levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of serum sodium levels")
plt.xlabel("serum sodium levels")
plt.ylabel("Count")
plt.show()
```
