```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Analysis of Female Survivors

```python
femaleSurvivors = pd.read_csv("femaleSurvivors.csv")
```

```python
femaleSurvivors
```

```python
femaleSurvivors.describe()
```

```python
femaleSurvivors["CPK Levels"] = pd.cut(x=femaleSurvivors["creatinine_phosphokinase"],bins = [10,120,7861], labels=["normal","high"])
```

```python
femaleSurvivors["ejection fraction levels"] = pd.cut(x=femaleSurvivors["ejection_fraction"],bins = [0,30,40,51,72], labels=["severely abnormal","moderately abnormal","mildly abnormal","normal"])
```

```python
femaleSurvivors["serum creatinine levels"] = pd.cut(x=femaleSurvivors["serum_creatinine"],bins = [0,0.7,1.3,6.1], labels=["low","normal","abnormal"])
```

```python
femaleSurvivors["serum sodium levels"] = pd.cut(x=femaleSurvivors["serum_sodium"],bins = [0,135,145,148], labels=["low","normal","abnormal"])
```

```python
counts = femaleSurvivors['CPK Levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of CPK Levels")
plt.xlabel("CPK Levels")
plt.ylabel("Count")
plt.show()
```

```python
counts = femaleSurvivors['ejection fraction levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of ejection fraction levels")
plt.xlabel("ejection fraction levels")
plt.ylabel("Count")
plt.show()
```

```python
counts = femaleSurvivors['serum creatinine levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of serum creatinine levels")
plt.xlabel("serum creatinine levels")
plt.ylabel("Count")
plt.show()
```

```python
counts = femaleSurvivors['serum sodium levels'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(counts.index, counts)
plt.title = ("Count Plot of serum sodium levels")
plt.xlabel("serum sodium levels")
plt.ylabel("Count")
plt.show()
```
