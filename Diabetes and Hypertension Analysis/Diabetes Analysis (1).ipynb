## Proportional Analysis

### Males

```python
diabetesTrueMS = (((maleSurvivor["diabetes"] == "TRUE").sum()) / len(maleSurvivor)) * 100
diabetesFalseMS = (((maleSurvivor["diabetes"] == "FALSE").sum())/ len(maleSurvivor)) * 100
```

```python
diabetesPercentSurvivor = {"DIABETES" : ["TRUE","FALSE"], "%" : [round(diabetesTrueMS,2), round(diabetesFalseMS,2)]}
diabetesPercentSDF = pd.DataFrame.from_dict(diabetesPercentSurvivor)
diabetesPercentSDF
```

```python
diabetesTrueMD = (((maleMortality["diabetes"] == "TRUE").sum()) / len(maleMortality)) * 100
diabetesFalseMD = (((maleMortality["diabetes"] == "FALSE").sum()) / len(maleMortality)) * 100
```

```python
diabetesPercentMortality = {"DIABETES" : ["TRUE", "FALSE"], "%" : [round(diabetesTrueMD, 2), round(diabetesFalseMD, 2)]}
diabetesPercentMDF = pd.DataFrame.from_dict(diabetesPercentMortality)
diabetesPercentMDF
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])


```

```python
x = diabetesPercentSDF["DIABETES"]
y = diabetesPercentSDF["%"]


diabetesPercentSDF.plot(x = "DIABETES", y = "%", kind = "bar", ylim = (0,100))
addlabels(x,y)
plt.grid()
```

```python
x = diabetesPercentMDF["DIABETES"]
y = diabetesPercentMDF["%"]


diabetesPercentMDF.plot(x = "DIABETES", y = "%", kind = "bar", ylim = (0,100), color = "red")
addlabels(x,y)
plt.grid()
```

```python
survivors = pd.read_csv("survivors.csv")
mortality = pd.read_csv("deaths.csv")
survivors["diabetes"].replace(0,"FALSE", inplace = True)
survivors["diabetes"].replace(1, "TRUE", inplace = True)
mortality["diabetes"].replace(0,"FALSE", inplace = True)
mortality["diabetes"].replace(1, "TRUE", inplace = True)
```

```python
survivors["diabetes"].value_counts()
```

```python
mortality["diabetes"].value_counts()
```

```python
counts = survivors.groupby(['diabetes', 'sex']).size().unstack(fill_value=0)
index = range(len(counts))
bar_width = 0.35
fig, ax = plt.subplots()
bars1 = plt.bar(index, counts[1], bar_width, label='Male', color='blue')
bars2 = plt.bar([i + bar_width for i in index], counts[0], bar_width, label='Female', color='pink')
ax.set_title = ("Count Plot of Diabetes Levels for Survivors")
plt.xlabel("Diabetes")
ax.set_ylim(0,len(deaths))
plt.ylabel("Count")

for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    max_height = max(bar1.get_height(), bar2.get_height())
    ejection_fraction_level = counts.index[i]
    plt.text(bar1.get_x() + bar1.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center', weight = "bold", color = "green")
    plt.text(bar2.get_x() + bar2.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center', weight = "bold", color = "green")

plt.grid()
plt.show()
```
