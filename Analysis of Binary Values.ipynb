```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Analysis for the male overall

```python

maledf = pd.read_csv("Male.csv")
```

```python
maledf.describe()
```

### Analysis for binary features

#### Diabetes

```python

diabetes_counts = maledf["diabetes"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(diabetes_counts.index, diabetes_counts)
plt.title("Analysis of Diabetes for Males")
plt.xlabel("Diabetes for Males")
plt.ylabel("Count")
```

High Blood Pressure

```python
blood_pressure_counts = maledf["high_blood_pressure"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(blood_pressure_counts.index, blood_pressure_counts)
plt.title("Analysis of High Blood Pressure for Males")
plt.xlabel("High_Blood_pressure for Males")
plt.ylabel("Count")
```

```python
anemia_counts = maledf["anaemia"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(anemia_counts.index, anemia_counts)
plt.title("Analysis of anemia for Males")
plt.xlabel("Anemia for Males")
plt.ylabel("Count")
```

```python
smoking_counts = maledf["smoking"].value_counts()

plt.figure(figsize=(8,6))
plt.bar(smoking_counts.index, smoking_counts)
plt.title("Analysis of Smoking for Males")
plt.xlabel("Smoking for Males")
plt.ylabel("Count")
```
