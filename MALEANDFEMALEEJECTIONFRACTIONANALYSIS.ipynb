```python
survivors = pd.read_csv("survivors.csv")
deaths = pd.read_csv("deaths.csv")
```

```python
len(survivors)
```

```python
len(deaths)
```

```python
survivors["Ejection_Fraction_Levels"] = pd.cut(x=survivors["ejection_fraction"], bins = [0,40,49,survivors["ejection_fraction"].max()], labels = ["reduced","preserved","normal"])
deaths["Ejection_Fraction_Levels"] = pd.cut(x=deaths["ejection_fraction"], bins = [0,40,49,deaths["ejection_fraction"].max()], labels = ["reduced","preserved","normal"])
```

```python
survivors["Ejection_Fraction_Levels"]
```

```python
deaths["Ejection_Fraction_Levels"]
```

```python
counts = survivors.groupby(['Ejection_Fraction_Levels', 'sex']).size().unstack(fill_value=0)
index = range(len(counts))
bar_width = 0.35
fig, ax = plt.subplots()
bars1 = plt.bar(index, counts[1], bar_width, label='Male', color='blue')
bars2 = plt.bar([i + bar_width for i in index], counts[0], bar_width, label='Female', color='pink')
ax.set_title = ("Count Plot of Ejection Fraction Levels for Survivors")
plt.xlabel("Ejection Fraction Levels")
ax.set_ylim(0,len(deaths))
plt.ylabel("Count")

for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    max_height = max(bar1.get_height(), bar2.get_height())
    ejection_fraction_level = counts.index[i]
    plt.text(bar1.get_x() + bar1.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center')
    plt.text(bar2.get_x() + bar2.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center')

plt.grid()
plt.show()
```

```python
counts = deaths.groupby(['Ejection_Fraction_Levels', 'sex']).size().unstack(fill_value=0)
index = range(len(counts))
bar_width = 0.35
fig, ax = plt.subplots()
bars1 = plt.bar(index, counts[1], bar_width, label='Male', color='blue')
bars2 = plt.bar([i + bar_width for i in index], counts[0], bar_width, label='Female', color='pink')
ax.set_title = ("Count Plot of Ejection Fraction Levels for mortality")
plt.xlabel("Ejection Fraction Levels")
ax.set_ylim(0,len(deaths))
plt.ylabel("Count")

for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, int(yval), ha='center', va='bottom')

for i, (bar1, bar2) in enumerate(zip(bars1, bars2)):
    max_height = max(bar1.get_height(), bar2.get_height())
    ejection_fraction_level = counts.index[i]
    plt.text(bar1.get_x() + bar1.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center')
    plt.text(bar2.get_x() + bar2.get_width() / 2, -0.5, ejection_fraction_level, ha='center', va='center')

plt.grid()
plt.show()
```

## Proportional Analysis


```python
reduced_percentageMS = (((maleS["Ejection Fraction Levels"] == "reduced").sum()) / len(maleS)) * 100
normal_percentageMS = (((maleS["Ejection Fraction Levels"] == "normal").sum()) / len(maleS)) * 100
preserved_percentageMS = (((maleS["Ejection Fraction Levels"] == "preserved").sum()) / len(maleS)) * 100
```

```python
reduced_percentageMD = (((maleD["Ejection Fraction Levels"] == "reduced").sum()) / len(maleD)) * 100
normal_percentageMD = (((maleD["Ejection Fraction Levels"] == "normal").sum()) / len(maleD)) * 100
preserved_percentageMD = (((maleD["Ejection Fraction Levels"] == "preserved").sum()) / len(maleD)) * 100
```

```python
maleSurvivorEF = {"EF_LEVELS" : ["REDUCED", "PRESERVED","NORMAL"], "%" : [round(reduced_percentageMS, 2), round(preserved_percentageMS, 2),round(normal_percentageMS,2)] }
maleMortalityEF = {"EF_LEVELS" : ["REDUCED", "PRESERVED", "NORMAL"], "%" : [round(reduced_percentageMD, 2), round(preserved_percentageMD, 2), round(normal_percentageMD,2)]}
```

```python
maleSurvivorEFDF = pd.DataFrame.from_dict(maleSurvivorEF)
maleMortalityEFDF = pd.DataFrame.from_dict(maleMortalityEF)
```

```python
maleSurvivorEFDF
```

```python
maleMortalityEFDF
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

x = maleSurvivorEFDF["EF_LEVELS"]
y = maleSurvivorEFDF["%"]


maleSurvivorEFDF.plot(x = "EF_LEVELS", y = "%", kind = "bar", ylim = (0,100))
addlabels(x,y)
plt.grid()

```

```python
x = maleMortalityEFDF["EF_LEVELS"]
y = maleMortalityEFDF["%"]
maleMortalityEFDF.plot(x = "EF_LEVELS", y = "%", kind = "bar", ylim = (0,100))
addlabels(x,y)
plt.grid()
```
